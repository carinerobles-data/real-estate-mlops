```python
import pandas as pd
import numpy as np
import os

# Configuration
RAW_DATA_PATH = "../data/raw/full2025.csv"
OUTPUT_PATH = "../data/processed/dvf_load.parquet"

cols_to_keep = [
    "id_mutation", "date_mutation", "nature_mutation", "valeur_fonciere",
    "code_departement", "code_commune", "nom_commune",
    "adresse_nom_voie", "type_local", 
    "surface_reelle_bati", "nombre_pieces_principales", "surface_terrain",
    "longitude", "latitude"
]

# --- VÉRIFICATION ET CHARGEMENT ---
try:
    # 1. Vérification des colonnes
    header = pd.read_csv(RAW_DATA_PATH, nrows=0)
    missing_cols = [c for c in cols_to_keep if c not in header.columns]
    
    if missing_cols:
        print(f"ALERTE : Colonnes manquantes : {missing_cols}")
    else:
        # 2. Chargement
        df = pd.read_csv(RAW_DATA_PATH, sep=',', usecols=cols_to_keep, low_memory=False)
        
        # 3. Audit des données
        print("--- 5 Premières lignes ---")
        display(df.head())
        
        print("\n--- Informations sur les colonnes (Types et Non-Nulls) ---")
        display(df.info())
        
        # 4. Export
        df.to_parquet(OUTPUT_PATH, index=False)
        print(f"\nSuccès : Fichier sauvegardé sous {OUTPUT_PATH}")

except FileNotFoundError:
    print(f"Erreur : Le fichier {RAW_DATA_PATH} est introuvable.")
```
